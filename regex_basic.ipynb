# Regex Basic

```python
#Define a simple pattern
pattern = r'\d+' #Matches one or more digits

# Create a Regex object using re.compile()
regex_object = re.compile(pattern)
```

```python
text = "There are 100 mangos and 456 bananas"

match = re.search(regex_object,text) #search ressort le premier match
# match = regex_object.search(text)
if match:
    print('Pattern found : ', match.group())
else :
    print('Pattern not found')
```

```python
text = "There are 100 mangos and 456 bananas"

match = re.findall(regex_object,text) #findall ressort un array de tous les match
# match = regex_object.search(text)

if match:
    print('Pattern found : ', match)
else :
    print('Pattern not found')
```

re.match():

The re.match(pattern, string) function in Python's re module is used to determine if a given regular expression pattern matches at the beginning of a string. It returns a match object if the pattern is found, otherwise, it returns None. Here's an example of how to use re.match():

```python
# Target String
target_string = "Hello, world!"

# match() for 'Hello' at the start of the string
result = re.match(r'^Hello', target_string)

# Print match object
print("Match Object", result)

# print the matching word using group() method
print("\nMatching word: ", result.group())
```

The ^ character is a metacharacter that represents the start of a line

Let's consider an simple example:

```python
# Define a pattern
pattern = r'apple'  # Matches the word 'apple'

# Sample text
text = 'I have an apple and a banana'

# Using re.match() to match the pattern at the beginning of the string
match = re.match(pattern, text) #search trouverait un match

if match:
    print('Pattern found:', match.group())
else:
    print('Pattern not found')
```

```python
# Target String
target_string = "Hello, world!"

# match() for 'world' at the end of the string
result = re.match(r'world!$', target_string)

# Print match object
print("Match Object: ", result)

# print the matching word using group() method
if result:
  print("Matching word: ", result.group())
else:
  print("No match found.")
```

The $ character is a metacharacter that represents the end of a line. If the pattern is found at the end of the string, the match object’s group() method will return the matched text (“world!” in this case)

The match() function returns a Match object only if it finds a pattern at the beginning of a string. For example:

```python
l = ['Python',
    'CPython is an implementation of Python written in C',
    'Jython is a Java implementation of Python',
     'IronPython is Python on .NET framework']

pattern = r'\wython'
for s in l:
    result = re.match(pattern,s)
    print(result)
```

Here's another example demonstrating the usage of re.match() with a compiled regex object:

```python
# Define a pattern
pattern = r'\d+'  # Matches one or more digits

# Create a regex object using re.compile()
regex_object = re.compile(pattern)

# Sample text
text = '123abc456def'

# Using re.match() with the compiled regex object
match = regex_object.match(text)

if match:
    print('Pattern found:', match.group())
else:
    print('Pattern not found at the beginning of the text')
```

```python
print(match.start())
print(match.end())
print(match.span())
```

```python
log = "2022-02-27 08:12:45 INFO This is a log message."

pattern = r"(\d{4}-\d{2}-\d{2}) (\d{2}:\d{2}:\d{2}) ([A-Z]+) (.+)"
match = re.match(pattern, log)

if match:
    print("Date:", match.group(1))
    print("Time:", match.group(2))
    print("Level:", match.group(3))
    print("Message:", match.group(4))
```

match.start(): This method returns the starting index of the matched text within the original string. It gives the position where the match starts in the string.

match.end(): This method returns the ending index of the matched text within the original string. It gives the position where the match ends in the string.

match.span(): This method combines the functionality of match.start() and match.end() by returning a tuple containing both the start and end index positions of the matched substring. The tuple represents a span indicating the range of indices where the match occurred within the text.

```python
pattern = r'\b\d+\b'
text = 'The numbers are 42, 1001, and 7.'

matches = re.finditer(pattern, text)
for match in matches:
   print("Matched text:", match.group())
   print("Start index:", match.start())
   print("End index:", match.end())
   print("Indices:", match.span())
   print()
```
