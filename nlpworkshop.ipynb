```python
def standardize_accented_chars(text):
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
```

# 4. Noise Removal

```python
# 1. Removing HTML tag

def remove_html_tags_regex(text):
    #remove html tags using regex
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
html_text = "<p>Hello, <b>world</b>!</p> <a href ='#' >Click here</a>"
cleaned_html = remove_html_tags_regex(html_text)
print(cleaned_html)
```

```python
# 2. Removing numbers/digits
def replace_numbers(text):
    result = re.sub(r'\d+', '<phone number>', text)
    return result
```

```python
input_numbers_text = "Her phone 554444 number is 0123456789, and my phone number is 0987654321"
replace_numbers(input_numbers_text)
```

```python
# 3. Convert the numbers into words
import inflect 
p = inflect.engine()
```

```python
def convert_number(text):
    #split string into list of words
    temp_str = text.split()
    # initialize empty list
    new_string = []
    
    for word in temp_str:
        # if word is a digit, convert the digit to the word in letters and append into the new_string list
        if word.isdigit():
            temp = p.number_to_words(word)
            new_string.append(temp)
            
        # append the word as it is 
        else:
            new_string.append(word)
            
    #join the words of new_string to form a string
    temp_str = ' '.join(new_string)
    return temp_str
```

```python
input = "There are 3 balls in this bag, and 12 in the other one."
convert_number(input)
```

```python
# 4. Removing punctuation & special chars
regular_punct = list(string.punctuation)
print(regular_punct)
```

```python
def remove_punctuation(text, punct_list):
    for punc in punct_list :
        if punc in text:
            text = text.replace(punc,'')
    return text.strip()
```

```python
remove_punctuation("Hey, @Lisa, did you know that the summer break is coming? Amazing right!! It's only 5 more days !!", regular_punct)
```

```python
# 5. Handling double whitespace from text

def remove_whitespace(text):
    return " ".join(text.split())

input_ws = "we don't need the   given     questions."
remove_whitespace(input_ws)
```

```python
# 6. Handling emoji/emoticon ---> for sentiment analysis, emoji/emoticon is usually not removed

def remove_emojis_regex(text):
    # remove emojis using Unicode emoji ranges.
    # Unicode ranges for emojis
   emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
    u"\U00002702-\U000027B0"  # dingbats
    u"\U000024C2-\U0001F251"
    u"\U0001f926-\U0001f937"
    u"\U00010000-\U0010ffff"
    "]+", flags=re.UNICODE)
   
   return emoji_pattern.sub(r'', text)
```

```python
emoji_text = "Hello! 😃 How are you ? 🎉 I'm in NYC 🇺🇸"

cleaned_emoji = remove_emojis_regex(emoji_text)
cleaned_emoji
```

```python
# 7. Removal of URLs or other irrelevant hints

#Define a regex pattern to match URLs
url_pattern = re.compile(r'https?://\S+')

#Define a function to remove URLs from text
def remove_urls(text):
    return url_pattern.sub('', text) # we remove the url

text_url = "Visit https://www.google.com for more information."
remove_urls(text_url)
    
```

```python
url_pattern = []
url_pattern.append(re.compile(r'https?://\S+'))
url_pattern.append(re.compile(r'http?://\S+'))
url_pattern.append(re.compile(r'www.\S+'))

def replace_more_urls(text):
    for pattern in url_pattern:
        text= pattern.sub('<this link>', text) # we replace the url with <this link>
    return text

more_url_text = "Visit www.google.com for info"
replace_more_urls(more_url_text)
```

# 6. Stopwords removal

Stopwords are words that do not contribute to the meaning of a sentence. Hence, they can safely be removed without causing any change in the meaning 
of the sentence. The NLTK library has a set of stopwords and we can use these to remove stopwrods from our text and return a list of word tokens

```python
# Stopwords removal
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
# Customize the stopwords set
stop_words.add('test')
print(stop_words)
```

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = nltk.word_tokenize(text)
    
    #filter out the stopwords
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    
    return filtered_text
```

```python
example_text = "This is a sample sentence and we are going to remove the stopwords from this."
print (f"Before remowing stopwords : {example_text}")
print (f"Aftere removing stopwords : {remove_stopwords(example_text)}")
```

# 7. Stemming & Lemmatization

Various algorithms form stemming : <br>
- Porter Stemmer <br>
- Snowball Stemmer <br>
- Lovins Stemmer

```python
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer = PorterStemmer()

#stem words in the list of tokenized words
def stem_words(text):
    word_tokens = word_tokenize(text=text)
    stems = [stemmer.stem(word) for word in word_tokens]
    return stems

stem_text = "data science uses scientific methods algorithms and many types of processes"
stem_words(stem_text)

```

```python
# Lemmatzation using NLTK
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# Function to lemmatize words in the list of tokenized words
def lemma_words(text):
    word_tokens = word_tokenize(text)
    lemmas = [lemmatizer.lemmatize(word) for word in word_tokens]
    return lemmas

text = 'The quick brown foxes are jumping over the lazy dogs.'
lemma_words(text)
```
